<a id="importacaoBibliotecas"></a>
Importa√ß√£o das bibliotecas Pandas e Plotly.

```python
import pandas as pd
import plotly.express as px
```

<a id="importacaoDadoCenso"></a>
##Importa√ß√£o dos dados do GitHub e defin√ß√£o do Banco de Dados.


```python

df = pd.DataFrame(pd.read_csv('https://github.com/janosystime/Janosys-Project/blob/main/bd/mockado_df.csv?raw=true', encoding='latin1', sep=';'))
```

<a id="displayDF"></a>
##Exibi√ß√£o do DataFrame Bruto e informa√ß√£o das colunas.

```python
display(df.head())
display(df.info())
```

<a id="filtragemDF"></a>
##Filtragem e exibi√ß√£o do Dataframe por Regi√£o e Faixa-Et√°ria

```python
totalRegioes_df = df[['Regiao', 'CENSO 2010', 'CENSO 2022']]

display(totalRegioes_df.head())
```

<a id="definicaoPlotly"></a>
##Definindo gr√°fico no plotly com base nos filtros escolhidos.

```python
fig = px.bar(totalRegioes_df, x='Regiao', y=['CENSO 2010', 'CENSO 2022'], title='Popula√ß√£o por Regi√£o (2010 vs 2022)',
labels={'Diferen√ßa': 'Diferen√ßa de Popula√ß√£o', 'Regiao': 'Regi√£o'}, barmode='group', text_auto='.2s', color_discrete_sequence=['#A8C8FF','#FFF5B1'])
fig.update_traces(textfont_size=12, textangle=0, textposition="outside", cliponaxis=False)
```

<a id="displayPlotly"></a>
##Exibi√ß√£o do gr√°fico gerado.

```python

fig.show()
```
